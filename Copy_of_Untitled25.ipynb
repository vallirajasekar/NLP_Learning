```python
from transformers import MarianMTModel, MarianTokenizer
```

```python
model_name = "Helsinki-NLP/opus-mt-de-en"
model = MarianMTModel.from_pretrained(model_name)
tokenizer = MarianTokenizer.from_pretrained(model_name)
```

```python
input_text_german = "Hallo, wie geht es Ihnen?"
```

```python
input_ids_german = tokenizer(input_text_german, return_tensors="pt", truncation=True)['input_ids']
output_ids_english = model.generate(input_ids_german)
translated_text_english = tokenizer.decode(output_ids_english[0], skip_special_tokens=True)
```

```python
# Print the translated text
print("Input Text (German):", input_text_german)
print("Translated Text (English):", translated_text_english)
```

```python
import pandas as pd
```

```python
# Load pre-trained translation model and tokenizer for German to English
model_name = "Helsinki-NLP/opus-mt-de-en"
model = MarianMTModel.from_pretrained(model_name)
tokenizer = MarianTokenizer.from_pretrained(model_name)
```

```python
data = {'German_Text': ['Hallo, wie geht es Ihnen?', 'Ich liebe maschinelles Lernen.', 'Guten Morgen!']}
df = pd.DataFrame(data)
```

```python
def translate_to_english(german_text):
    input_ids_german = tokenizer(german_text, return_tensors="pt", truncation=True)['input_ids']
    output_ids_english = model.generate(input_ids_german)
    translated_text_english = tokenizer.decode(output_ids_english[0], skip_special_tokens=True)
    return translated_text_english
```

```python
df['English_Text'] = df['German_Text'].apply(translate_to_english)
```
